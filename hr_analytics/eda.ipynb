```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.shape
```

```python
attrition_rate = round((237 / 1470) * 100, 2)
print(f"Attrition Rate: {attrition_rate}%")
```

```python
sns.countplot(x='Attrition', data=df, palette=['#2ecc71','#e74c3c'])
plt.title('Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()['Yes'].sort_values(ascending=False)
```

```python
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()['Yes'].sort_values(ascending=False)

sns.barplot(x=dept_attrition.index, y=dept_attrition.values, palette='Reds_r')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()
```

```python
overtime_attrition = df.groupby('OverTime')['Attrition'].value_counts(normalize=True).unstack()['Yes']

sns.barplot(x=overtime_attrition.index, y=overtime_attrition.values, palette=['#2ecc71','#e74c3c'])
plt.title('Attrition Rate by Overtime')
plt.xlabel('Overtime')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()

print(overtime_attrition)
```

```python
satisfaction_attrition = df.groupby('JobSatisfaction')['Attrition'].value_counts(normalize=True).unstack()['Yes']

sns.barplot(x=satisfaction_attrition.index, y=satisfaction_attrition.values, palette='Oranges')
plt.title('Attrition Rate by Job Satisfaction')
plt.xlabel('Job Satisfaction (1=Low, 4=High)')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()

print(satisfaction_attrition)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20, palette=['#2ecc71','#e74c3c'])
plt.title('Attrition by Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[18,25,35,45,60], labels=['18-25','26-35','36-45','46-60'])

age_attrition = df.groupby('AgeGroup')['Attrition'].value_counts(normalize=True).unstack()['Yes']

sns.barplot(x=age_attrition.index, y=age_attrition.values, palette='Blues_r')
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()

print(age_attrition)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette=['#2ecc71','#e74c3c'])
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()

print(df.groupby('Attrition')['MonthlyIncome'].mean().round(0))
```
